```python
import pandas as pd
import glob

path="G:/QTI/NER_project/NER -Tool/ner-tagger-tool/st_ner_annotate/backend/entity/"
all_files = os.listdir(path)

word_entity = []

for csvname in all_files:
    df = pd.read_csv(path+csvname, index_col=None, header=0)
    word_entity.append(df)
    
frame = pd.concat(word_entity, axis=0, ignore_index=True)
```

```python
tulis = frame.values.tolist()
list_token = []
for count in range(len(tulis)):
    list_token.append(tulis[count][0])
    
print(list_token)
```

```python
list_predict = []
for word in list_token:
    token = model_path(str(word))
    list_predict.append(token)
```

```python
has_vector = []
for token in list_predict:
    for ent in token.ents:
        print(ent.text, ent.label_)
    print(token.has_vector)
    print("")
    if token.has_vector==True:
        has_vector.append(token)
print(has_vector)
```
