```python
import csv
from pathlib import Path
```

```python
def get_qid():
    ent_loc = Path.cwd()/"wikidata"/"indonesian_people.csv"
    list_qid = []
    with ent_loc.open("r", encoding="utf8") as csvfile:
        csvreader = csv.reader(csvfile, delimiter=",")
        for row in csvreader:
            list_qid.append(row[0])
        list_qid.remove("item")
        list_qid = [id.split("/")[-1] for id in list_qid]
    return list_qid
```

```python
def load_entities():
    ent_loc = Path.cwd()/"wikidata"/"indonesian_people.csv"
    names = dict()
    description = dict()
    list_qid = get_qid()
    with ent_loc.open("r", encoding="utf8") as csvfile:
        csvreader = csv.reader(csvfile, delimiter=",")
        next(csvreader) #skip the header of csv file
        for index,row in enumerate(csvreader):
            qid = list_qid[index]
            name = row[1]
            desc = row[2]
            names[qid] = name
            description[qid] = desc
    return names,description
```

```python
name_dict, desc_dict = load_entities()
```

```python
with open("name_csv.csv", "w", encoding="utf8") as csvfile:
    writer = csv.writer(csvfile)
    for key, val in name_dict.items():
        writer.writerow([key, val, desc_dict[key]])
```
